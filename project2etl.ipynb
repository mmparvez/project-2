```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
import config
```

### EXTRACT
## Store CSV into DataFrame

```python
#Read 1st csv files as pandas dataframe and view
csv_file = "Raw Data/CO2 Emission Per Cap.csv"
co2_df = pd.read_csv(csv_file, skiprows=3)
co2_df.head()
```

```python
#Read 2nd csv files as pandas dataframe and view
csv_file = "Raw Data/GDP Per Cap.csv"
gdp_df = pd.read_csv(csv_file, skiprows=3)
gdp_df.head()
```

### TRANSFORM
## Clean DataFrame

```python
#Split 1st dataframe to drop columns that are not needed
#1st dataframe
#First part of the split
co2_1st_df =co2_df.iloc[:,0:2]
co2_1st_df.head()
```

```python
#1st dataframe
#Second part of the split
#Droped columns that are not needed
co2_2nd_df = co2_df.loc[:, '1990':'2021']
co2_2nd_df.head()
```

```python
#Split 2nd dataframe to drop columns that are not needed
#2nd dataframe
#First part of the split
gdp_1st_df = gdp_df.iloc[:,0:2]
gdp_1st_df.head()
```

```python
#2nd dataframe
#Droped columns that are not needed
gdp_2nd_df = gdp_df.loc[:, '1990':'2021']
gdp_2nd_df.head()
```

```python
#Concatenate 2nd dataframe
gdp_new = [gdp_1st_df, gdp_2nd_df]
gdp_final_df = pd.concat(gdp_new, axis=1)
#Rename columns of 2nd dataframe
gdp_final_df = gdp_final_df.rename(columns={"Country Name": "country_name", "Country Code": "country_code"\
     , "1990": "gdp_1990", "1991": "gdp_1991", "1992": "gdp_1992", "1993": "gdp_1993", "1994": "gdp_1994", \
         "1995": "gdp_1995", "1996": "gdp_1996", "1997": "gdp_1997", "1998": "gdp_1998", "1999": "gdp_1999", \
            "2000": "gdp_2000", "2001": "gdp_2001", "2002": "gdp_2002", "2003": "gdp_2003", "2004": "gdp_2004", \
                 "2005": "gdp_2005", "2006": "gdp_2006", "2007": "gdp_2007", "2008": "gdp_2008", "2009": "gdp_2009", \
                     "2010": "gdp_2010", "2011": "gdp_2011", "2012": "gdp_2012", "2013": "gdp_2013", "2014": "gdp_2014", \
                         "2015": "gdp_2015", "2016": "gdp_2016", "2017": "gdp_2017", "2018": "gdp_2018", "2019": "gdp_2019", \
                             "2020": "gdp_2020", "2021": "gdp_2021"})
gdp_final_df.head()
```

### LOAD
## Connect to local database

```python
#Connect to local SQL database
protocol = 'postgresql'
username = config.username #'<username>'
password = config.password #'<password>'
host = 'localhost'
port = 5432
database_name = 'project-2_DB'
#Create connection string
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
#Create engine
engine = create_engine(rds_connection_string)
insp = inspect(engine)
```

## Check for tables

```python
#Check for tables
insp.get_table_names()
```

## Use pandas to load csv converted DataFrame into database

```python
co2_final_df.to_sql(name='co2_emission', con=engine, if_exists='append', index=False)
```

## Use pandas to load json converted DataFrame into database

```python
gdp_final_df.to_sql(name='gdp_per_capita', con=engine, if_exists='append', index=False)
```
